```python
import requests
r=requests.get("https://storage.googleapis.com/laurencemoroney-blog.appspot.com/sarcasm.json")
open('./tmp/sarcasm.json',"wb").write(r.content)
```

```python
import json

with open("./tmp/sarcasm.json","r") as r:
    datastore=json.load(r)

sentences=[]
labels=[]
urls=[]
for data in datastore:
    sentences.append(data["headline"])
    labels.append(data["is_sarcastic"])
    urls.append(data["article_link"])
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
#Tokenizing the sentences 
tokenizer=Tokenizer(oov_token="<OOV>")
tokenizer.fit_on_texts(sentences)

word_index=tokenizer.word_index
print(len(word_index))
print(word_index)
sequences=tokenizer.texts_to_sequences(sentences)
padded=pad_sequences(sequences,padding="post")
```
